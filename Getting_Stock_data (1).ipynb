```
from pandas_datareader import data
# only works on Google Colab, not locally

import matplotlib.pyplot as plt
import pandas as pd
import datetime
import os
```

```
# Google Colab's Datareader

tickers = ['AAPL', 'MSFT', 'FB', 'GOOG', 'AMZN', '^GSPC', '^IXIC']

start_date = datetime.datetime(2020,9,4)
end_date = datetime.datetime(2020,12,3)

panel_data = data.DataReader(tickers, 'yahoo', start_date, end_date)

df = pd.DataFrame(panel_data)
```

```
df['Close']
```

```
close = df['Close']
close = pd.DataFrame(close)
```

```
# closing stock price for 2020-10-30 to 2020-11-25
close.drop(close.tail(1).index,inplace=True)
```

```
close
```

```
open = df['Open']
open = pd.DataFrame(open)
```

```
# opening stock price for 2020-11-02 to 2020-11-27
open.drop(open.head(1).index,inplace=True)
```

```
open
```

```
diff = open.subtract(close.values)
```

```
# difference = opening price - closing price the day before
diff
```

```

diff.to_csv('diff_3mo_final.csv', mode="a")
```
